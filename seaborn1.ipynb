# Pair plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"], hue="sex",hue_order=["Female","Male"]
             ,palette='BuGn')
plt.show()
```

```python
sns.pairplot(var,x_vars=["total_bill","tip"], hue="sex",hue_order=["Female","Male"]
             ,palette='BuGn')
plt.show()
```

```python
sns.pairplot(var,y_vars=["total_bill","tip"], hue="sex",hue_order=["Female","Male"]
             )
plt.show()
```

```python
sns.pairplot(var, hue="sex",hue_order=["Female","Male"],kind="reg")
plt.show()
```

```python
sns.pairplot(var, hue="sex",hue_order=["Female","Male"],kind="kde",diag_kind="hist",markers=["*","<"])
plt.show()
```
